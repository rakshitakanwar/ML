```python
#2-d array
import numpy as np
a = np.array([[10,20,30],[4,5,6]])
print(a.shape) # 2 rows and 3 columns
print(a.size) #no of elements
print(a.dtype) #data type
print(type(a)) #N-dimensional array
```
