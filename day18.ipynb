```python
n=1224
num=n
result=0
while num>0:
  last= num%10
  result= result*10+last
  num=num//10
if n==result:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
#print each character with its frequency
a= input("enter a string:")
freq={}
for ch in a:
  if ch in freq:
    freq[ch]+=1
  else:
    freq[ch]=1
print(freq)
```

```python
def rev(s):
  letter=[]
  for ch in s:
    if ch.isalpha():
      letter.append(ch)
  result=""
  for ch in s:
    if ch.isalpha():
      result+=letter.pop()
    else:
      result+=ch
  return result
print(rev("ab-cd"))
```

```python
s = "aababbbccccc"

count = 1

for i in range(len(s)-1):

    if s[i] == s[i+1]:
        count += 1

    else:
        print(s[i], count,sep="", end="")
        count = 1

print(s[-1], count, sep="")
```
